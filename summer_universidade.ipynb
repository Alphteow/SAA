```python
def convert(df):
    l=len(df)
    string='Rank'

    all_results = pd.DataFrame()


    for i in range(0,l):
        if string in df[i]:

            df[i]['Stage']=np.nan
            df[i]['Year']=np.nan
            df[i]['Event']=np.nan
            if df[i]['Notes'].str.contains('Q').any():
                df[i]['Stage']='Qualification'
                
            else:
                df[i]['Stage']='Final'
        
            all_results=pd.concat([all_results, df[i].copy()])
            
    string=df[0].columns[0]

    x=string.split(' at')

    y=x[1].split('the ')
    
    all_results['Event']=x[0]

    all_results['Year']=y[1]

            
    return all_results
```

```python
# Test integrated extraction of summer universidiade using soup for SEA Games
# Everything works except for heptahlon/decathlon events


#URL ='https://en.wikipedia.org/wiki/Athletics_at_the_2019_Summer_Universiade_%E2%80%93_Men%27s_100_metres'
#URL = 'https://en.wikipedia.org/wiki/Athletics_at_the_2019_Summer_Universiade_%E2%80%93_Men%27s_5000_metres'
#URL='https://en.wikipedia.org/wiki/Athletics_at_the_2019_Summer_Universiade_%E2%80%93_Men%27s_4_%C3%97_100_metres_relay'
#URL='https://en.wikipedia.org/wiki/Athletics_at_the_2019_Summer_Universiade_%E2%80%93_Men%27s_high_jump'
#URL='https://en.wikipedia.org/wiki/Athletics_at_the_2019_Summer_Universiade_%E2%80%93_Men%27s_shot_put'
URL='https://en.wikipedia.org/wiki/Athletics_at_the_2019_Summer_Universiade_%E2%80%93_Men%27s_javelin_throw'

page=requests.get(URL)


tables = pd.read_html(URL, match="Rank")  # extract tables only


table=[]

html_parsed=BeautifulSoup(page.content, 'html.parser')
soup = BeautifulSoup(str(html_parsed), 'html.parser')

year_pattern='\d\d\d\d'
competition_pattern='Summer Universiade'
gender_men_pattern='Men\'s'
gender_women_pattern='Women\'s'
event_pattern='\'s\s'
event_pattern2='\sat'
heat_pattern='Heat\s\d'
wind_pattern='(\+\d.\d\sm\/s|\-\d.\d\sm\/s|\-\d.\d\sm\/s|\d.\d\sm\/s)'
column_pattern='\d\.\d\d'  # to extract numeric columns for high jump, for example



df=pd.DataFrame(columns = ['gender', 'event', 'heat', 'results', 'country'])
list=[]
heat=None
event=None
results=None
country=None
what=None
result_list=[]
countries_list=[]
master_list=[]
heat_list=[]
list=[]

# Extract year, gender, competition and event details

event_a=soup.find('h1').find('span', class_='mw-page-title-main')
event_list=event_a.text

event=re.split(event_pattern2, re.split(event_pattern,event_list)[1])[0]

year=re.search(year_pattern, event_list)

event_year=year.group()

if competition_pattern in event_list:
    competition='Summer Universiade'
    
if gender_men_pattern in event_list:
    gender='Male'
else:
    gender='Female'

j=0   # counter for table number

for i in soup.find_all('h3'):
    
                
    round=i.find('span', class_='mw-headline') # stage of event. This is looping all the events downstream
            
    print(round.text)
    
    stage=round.text    
            
    for sib in i.next_siblings:   # h4>ul/li>table>td
                    
        if sib.name=='p' and 'Wind' in sib.text :    #Look for wind data
            
            wind=sib.text
                        
            Heat=re.findall(heat_pattern, wind) 
            Wind=re.findall(wind_pattern, wind)
            
            print(Heat)
            
            if Heat:  # if wind data is split into heats
            
                wind_data=dict(zip(Heat, Wind))  # create dict of wind speeds
                                                                
                tables[j]['test'] = "Heat " + tables[j].Heat.map(str)  # map heats to wind dictionary
                tables[j]['Wind']=tables[j]['test'].map(wind_data)
                
                
            else:
                
                if stage=='Final':
                    
                
                    match=re.search(wind_pattern, wind)
                    
                    tables[j]['Wind']=match[0]
            
            
        elif sib.name == 'table':   # results in tables
            
            # for putting distance event result tables into a dictionary
            
            if tables[j].filter(regex='\d\.\d\d|\#\d', axis=1).shape[1]!=0:  # check that distance tables are present
                                                
                dictionary_results=tables[j].filter(regex='\d\.\d\d|\#\d', axis=1) # extract column of results from distance events
                tables[j]['dict_results']=dictionary_results.to_dict('records') # put results into a dictionary
                tables[j].drop(tables[j].filter(regex='\d\.\d\d|\#\d').columns, axis=1, inplace=True) # drop redundant columns        
            
            tables[j]['Stage'] = stage
            tables[j]['Competition'] = competition
            tables[j]['Gender'] = gender
            tables[j]['Year'] = event_year
            tables[j]['Event'] = event
            
        elif sib.name == 'h3':  # get out and stop extracting siblings beyond the next h3 tag
        
            break
                         
                
    j+=1  # increment table counter
        
    
master_table=pd.DataFrame()
    
for i in range(len(tables)):      
    
                
    master_table=pd.concat([master_table, tables[i]])
            
            
if 'test' in master_table.columns:                

    master_table=master_table.drop(['test'], axis=1)  # skip if no wind data
                
```
