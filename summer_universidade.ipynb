```python
def convert(df):
    l=len(df)
    string='Rank'

    all_results = pd.DataFrame()


    for i in range(0,l):
        if string in df[i]:

            df[i]['Stage']=np.nan
            df[i]['Year']=np.nan
            df[i]['Event']=np.nan
            if df[i]['Notes'].str.contains('Q').any():
                df[i]['Stage']='Qualification'
                
            else:
                df[i]['Stage']='Final'
        
            all_results=pd.concat([all_results, df[i].copy()])
            
    string=df[0].columns[0]

    x=string.split(' at')

    y=x[1].split('the ')
    
    all_results['Event']=x[0]

    all_results['Year']=y[1]

            
    return all_results
```
